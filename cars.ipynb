```python
#Add all the data into lists to later create a dataframe
brand = []
model = []
category = []
year =[]
all_data = data["results"]
for x in range(len(all_data)):
    brand.append(all_data[x]["Make"])
    model.append(all_data[x]["Model"])
    category.append(all_data[x]["Category"])
    year.append(all_data[x]["Year"])
```

```python
#Create the Dataframe w the data
cars = pd.DataFrame({"Car": brand,
                    "Model": model,
                     "Category":category,
                     "Year":year
                    })
cars
```

```python
#Scrape the URl below to get a table
url = "https://www.goodcarbadcar.net/2020-us-vehicle-sales-figures-by-brand/"
web = requests.get(url)
soup = bs(web.text, 'html.parser')
results = soup.find('table', id="table_6")
results
```

```python
#After we scraped the table, we converted it into a csv file
#Then we import the csv file and add it to the code
table2 = pd.read_csv("tableproject.csv")
table2
```

```python
#Merging the two tables together 
finished_table= pd.merge(cars, table2, on ="Car", how="left")
pd.set_option("display.max_rows", None, "display.max_columns", None)

finished_table.dropna(how="any", inplace=True)

finished_table.reset_index(inplace=True)
finished_table.drop(columns = "index",inplace=True)
finished_table
```
